```python
data[['Position', 'Nationality', 'Club', 'Contract Valid Until']] = data[['Position', 'Nationality', 'Club', 'Contract Valid Until']]. fillna('none')
data['Wage'].replace(0,data['Wage'].mean(),inplace=True)
data['Value'] = data['Value']. fillna(data['Value'].mean())
data[['Release Clause']] = data[['Release Clause']]. fillna (35)
data[['Skill Moves', 'International Reputation',]] = data[['Skill Moves', 'International Reputation']]. fillna (1)
```

```python
data.describe()
```

```python
data = data.drop(['Joined','Contract Valid Until'],axis='columns') 
```

```python
data.info()
```

```python
data['Name'] = data['Name'].str.lower()
data['Nationality'] = data['Nationality'].str.lower()
data['Club'] = data['Club'].str.lower()
data['Preferred Foot'] = data['Preferred Foot'].str.lower()
data['Position'] = data['Position'].str.lower()
```

```python
data['Name'] = data['Name'].str.replace('\n', '')
data['Name'] = data['Name'].str.replace('\t', ' ')
data['Name'] = data['Name'].str.replace(' {2,}', ' ', regex=True)
data['Name'] = data['Name'].str.strip()
```

```python
data['Club'] = data['Club'].str.replace('1.', '', regex=True)
data['Club'] = data['Club'].str.replace('\n', '')
data['Club'] = data['Club'].str.replace('\t', ' ')
data['Club'] = data['Club'].str.replace(' {2,}', ' ', regex=True)
data['Club'] = data['Club'].str.strip()
```
